## 1) Environment Setup & Imports

```python
import os
import math
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.cuda.amp import autocast, GradScaler
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, DistributedSampler
import torch.multiprocessing as mp
import torch.distributed as dist
from torch.nn.parallel import DistributedDataParallel as DDP
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from PIL import Image
import warnings
warnings.filterwarnings('ignore')
```

```python
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
torch.backends.cudnn.benchmark = True

IMG_SIZE    = 224
PATCH_SIZE  = 16
NUM_PATCHES = (IMG_SIZE // PATCH_SIZE) ** 2       
MASK_RATIO  = 0.75
NUM_VISIBLE = int(NUM_PATCHES * (1 - MASK_RATIO)) 
NUM_MASKED  = NUM_PATCHES - NUM_VISIBLE            

ENC_DIM     = 768
ENC_LAYERS  = 12
ENC_HEADS   = 12

DEC_DIM     = 384
DEC_LAYERS  = 12
DEC_HEADS   = 6

BATCH_SIZE  = 64
EPOCHS      = 20
BASE_LR     = 1.5e-4
WEIGHT_DECAY= 0.05
WARMUP_EPOCHS = 10

DATA_PATH   = '/kaggle/input/datasets/akash2sharma/tiny-imagenet/tiny-imagenet-200/tiny-imagenet-200'
CKPT_DIR    = '/kaggle/working/checkpoints'
os.makedirs(CKPT_DIR, exist_ok=True)

print(f"Total patches: {NUM_PATCHES}")
print(f"Visible patches (25%): {NUM_VISIBLE}")
print(f"Masked patches (75%): {NUM_MASKED}")
```

## 2) Dataset Loading & DataLoader

```python
train_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(0.4, 0.4, 0.4, 0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = ImageFolder(os.path.join(DATA_PATH, 'train'), transform=train_transforms)
val_dataset   = ImageFolder(os.path.join(DATA_PATH, 'val'),   transform=val_transforms)

print(f"Train samples : {len(train_dataset)}")
print(f"Val samples   : {len(val_dataset)}")
print(f"Classes       : {len(train_dataset.classes)}")
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size  = BATCH_SIZE,
    shuffle     = True,
    num_workers = 4,
    pin_memory  = True,
    drop_last   = True
)

val_loader = DataLoader(
    val_dataset,
    batch_size  = BATCH_SIZE,
    shuffle     = False,
    num_workers = 4,
    pin_memory  = True,
    drop_last   = False
)

print(f"Train batches : {len(train_loader)}")
print(f"Val batches   : {len(val_loader)}")
```

## 3) Patchification & Masking

```python
class PatchEmbed(nn.Module):
    def __init__(self):
        super().__init__()
        self.proj = nn.Conv2d(3, ENC_DIM, kernel_size=PATCH_SIZE, stride=PATCH_SIZE)

    def forward(self, x):
        x = self.proj(x)
        x = x.flatten(2).transpose(1, 2)
        return x
```

```python
def get_2d_sincos_pos_embed(embed_dim, grid_size, cls_token=False):
    grid_h = np.arange(grid_size, dtype=np.float32)
    grid_w = np.arange(grid_size, dtype=np.float32)
    grid   = np.meshgrid(grid_w, grid_h)
    grid   = np.stack(grid, axis=0).reshape(2, -1)

    half   = embed_dim // 2
    omega  = np.arange(half // 2, dtype=np.float32)
    omega  = 1.0 / (10000 ** (2 * omega / half))

    out_h  = np.einsum('n,d->nd', grid[0], omega)
    out_w  = np.einsum('n,d->nd', grid[1], omega)

    emb_h  = np.concatenate([np.sin(out_h), np.cos(out_h)], axis=1)
    emb_w  = np.concatenate([np.sin(out_w), np.cos(out_w)], axis=1)

    emb = np.concatenate([emb_h, emb_w], axis=1)
    if cls_token:
        emb = np.concatenate([np.zeros((1, embed_dim)), emb], axis=0)

    return torch.tensor(emb, dtype=torch.float32)
```

```python
def random_masking(x, mask_ratio=MASK_RATIO):
    B, N, D   = x.shape
    keep      = int(N * (1 - mask_ratio))

    noise     = torch.rand(B, N, device=x.device)
    ids_shuffle  = torch.argsort(noise, dim=1)
    ids_restore  = torch.argsort(ids_shuffle, dim=1)

    ids_keep     = ids_shuffle[:, :keep]
    x_visible    = torch.gather(x, 1, ids_keep.unsqueeze(-1).expand(-1, -1, D))

    mask         = torch.ones(B, N, device=x.device)
    mask[:, :keep] = 0
    mask         = torch.gather(mask, 1, ids_restore)

    return x_visible, mask, ids_restore
```

```python
device     = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
patch_embed = PatchEmbed().to(device)
pos_embed   = get_2d_sincos_pos_embed(ENC_DIM, IMG_SIZE // PATCH_SIZE, cls_token=False).to(device)

sample, _  = next(iter(train_loader))
sample     = sample.to(device)

patches    = patch_embed(sample)
patches    = patches + pos_embed.unsqueeze(0)
x_vis, mask, ids_restore = random_masking(patches)

print(f"Input image shape      : {sample.shape}")
print(f"All patches shape      : {patches.shape}")
print(f"Visible patches shape  : {x_vis.shape}")
print(f"Mask shape             : {mask.shape}")
print(f"ids_restore shape      : {ids_restore.shape}")
print(f"Masked tokens (75%)    : {mask.sum(dim=1).mean():.0f}")
print(f"Visible tokens (25%)   : {(1-mask).sum(dim=1).mean():.0f}")
```

## 4) Encoder (ViT-Base)

```python
class TransformerBlock(nn.Module):
    def __init__(self, dim, num_heads, mlp_ratio=4.0, dropout=0.0):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn  = nn.MultiheadAttention(dim, num_heads, dropout=dropout, batch_first=True)
        self.norm2 = nn.LayerNorm(dim)
        self.mlp   = nn.Sequential(
            nn.Linear(dim, int(dim * mlp_ratio)),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(int(dim * mlp_ratio), dim),
            nn.Dropout(dropout)
        )

    def forward(self, x):
        normed = self.norm1(x)
        x = x + self.attn(normed, normed, normed, need_weights=False)[0]
        x = x + self.mlp(self.norm2(x))
        return x
```

```python
class MAEEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.patch_embed = PatchEmbed()
        
        # added cls_token
        self.cls_token = nn.Parameter(torch.zeros(1, 1, ENC_DIM))
        
        self.blocks      = nn.Sequential(*[TransformerBlock(ENC_DIM, ENC_HEADS) for _ in range(ENC_LAYERS)])
        self.norm        = nn.LayerNorm(ENC_DIM)

        pos = get_2d_sincos_pos_embed(ENC_DIM, IMG_SIZE // PATCH_SIZE, cls_token=True)
        self.register_buffer('pos_embed', pos.unsqueeze(0))

    def forward(self, x):
        x = self.patch_embed(x)
        
        # add pos_embed without cls token
        x = x + self.pos_embed[:, 1:, :]

        x, mask, ids_restore = random_masking(x)

        # append cls token
        cls_token = self.cls_token + self.pos_embed[:, :1, :]
        cls_tokens = cls_token.expand(x.shape[0], -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)

        x = self.blocks(x)
        x = self.norm(x)

        return x, mask, ids_restore
```

## 5) Decoder (ViT-Small)

```python
class MAEDecoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.proj        = nn.Linear(ENC_DIM, DEC_DIM)
        self.mask_token  = nn.Parameter(torch.zeros(1, 1, DEC_DIM))
        self.blocks      = nn.Sequential(*[TransformerBlock(DEC_DIM, DEC_HEADS) for _ in range(DEC_LAYERS)])
        self.norm        = nn.LayerNorm(DEC_DIM)
        self.pred        = nn.Linear(DEC_DIM, PATCH_SIZE * PATCH_SIZE * 3)

        pos = get_2d_sincos_pos_embed(DEC_DIM, IMG_SIZE // PATCH_SIZE, cls_token=True)
        self.register_buffer('pos_embed', pos.unsqueeze(0))

    def forward(self, x, ids_restore):
        x = self.proj(x)

        # separate cls token
        cls_token = x[:, :1, :]
        x_ = x[:, 1:, :]
        
        B, N, D  = x_.shape

        mask_tokens = self.mask_token.expand(B, ids_restore.shape[1] - N, -1)
        x_full      = torch.cat([x_, mask_tokens], dim=1)
        x_full      = torch.gather(x_full, 1, ids_restore.unsqueeze(-1).expand(-1, -1, DEC_DIM))

        # re-append cls token
        x_full   = torch.cat([cls_token, x_full], dim=1)

        x_full   = x_full + self.pos_embed
        x_full   = self.blocks(x_full)
        x_full   = self.norm(x_full)
        x_full   = self.pred(x_full)
        
        # return patches without cls token
        return x_full[:, 1:, :]
```

## 6) Full MAE Model + Forward Pass + Loss

```python
def patchify(imgs):
    B, C, H, W = imgs.shape
    p  = PATCH_SIZE
    h  = H // p
    x  = imgs.reshape(B, C, h, p, h, p)
    x  = x.permute(0, 2, 4, 3, 5, 1)
    x  = x.reshape(B, h * h, p * p * C)
    return x

def unpatchify(patches):
    B, N, _ = patches.shape
    p = PATCH_SIZE
    h = int(N ** 0.5)
    x = patches.reshape(B, h, h, p, p, 3)
    x = x.permute(0, 5, 1, 3, 2, 4)
    x = x.reshape(B, 3, h * p, h * p)
    return x
```

```python
class MAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = MAEEncoder()
        self.decoder = MAEDecoder()
        self.initialize_weights()
        
    def initialize_weights(self):
        # xavier_uniform init
        torch.nn.init.normal_(self.decoder.mask_token, std=0.02)
        torch.nn.init.normal_(self.encoder.cls_token, std=0.02)
        
        self.apply(self._init_weights)
        
    def _init_weights(self, m):
        if isinstance(m, nn.Linear):
            torch.nn.init.xavier_uniform_(m.weight)
            if isinstance(m, nn.Linear) and m.bias is not None:
                nn.init.constant_(m.bias, 0)
        elif isinstance(m, nn.LayerNorm):
            nn.init.constant_(m.bias, 0)
            nn.init.constant_(m.weight, 1.0)
        elif isinstance(m, nn.Conv2d):
            torch.nn.init.xavier_uniform_(m.weight)
            if m.bias is not None:
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        enc_out, mask, ids_restore = self.encoder(x)
        dec_out                    = self.decoder(enc_out, ids_restore)

        target = patchify(x)
        
        # patch normalization
        mean = target.mean(dim=-1, keepdim=True)
        var = target.var(dim=-1, unbiased=False, keepdim=True)
        target = (target - mean) / (var + 1.e-6)**.5
        
        loss   = F.mse_loss(dec_out[mask == 1], target[mask == 1])

        return loss, dec_out, mask
```

## 7) Training Loop

```python
device   = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
num_gpus = torch.cuda.device_count()

model    = MAE().to(device)

if num_gpus > 1:
    model = nn.DataParallel(model)

print(f"Training on : {num_gpus} GPU(s)")
print(f"Device      : {device}")
```

```python
param_groups = []
decay = []
no_decay = []
for name, param in model.named_parameters():
    if not param.requires_grad:
        continue
    if len(param.shape) == 1 or name.endswith(".bias") or "LayerNorm" in name or "pos_embed" in name or "cls_token" in name or "mask_token" in name:
        no_decay.append(param)
    else:
        decay.append(param)

param_groups = [
    {'params': decay, 'weight_decay': WEIGHT_DECAY},
    {'params': no_decay, 'weight_decay': 0.0}
]

optimizer  = torch.optim.AdamW(
    param_groups,
    lr           = BASE_LR * (BATCH_SIZE * num_gpus) / 256,
    betas        = (0.9, 0.95)
)

def cosine_lr_with_warmup(epoch):
    if epoch < WARMUP_EPOCHS:
        return max(epoch / max(WARMUP_EPOCHS, 1), 1e-6)
    progress = (epoch - WARMUP_EPOCHS) / max(EPOCHS - WARMUP_EPOCHS, 1)
    return max(0.5 * (1.0 + math.cos(math.pi * progress)), 1e-6)

scheduler  = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=cosine_lr_with_warmup)
scaler     = GradScaler()

print(f"PyTorch version : {torch.__version__}")
print(f"Effective LR    : {optimizer.param_groups[0]['lr']:.6f}")
print(f"Optimizer       : AdamW  |  Weight Decay : {WEIGHT_DECAY}")
print(f"Scheduler       : Cosine with {WARMUP_EPOCHS} epoch warmup")
```

```python
from tqdm import tqdm

train_losses = []
val_losses   = []
best_val     = float('inf')

for epoch in range(EPOCHS):
    model.train()
    epoch_loss = 0.0

    pbar = tqdm(train_loader, desc=f"Epoch [{epoch+1:>3}/{EPOCHS}]", ncols=100, leave=False)

    for imgs, _ in pbar:
        imgs = imgs.to(device, non_blocking=True)

        optimizer.zero_grad()
        with autocast():
            loss, _, _ = model(imgs)
            loss       = loss.mean()

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        scaler.step(optimizer)
        scaler.update()

        epoch_loss += loss.item()
        pbar.set_postfix(loss=f"{loss.item():.4f}", lr=f"{scheduler.get_last_lr()[0]:.6f}")

    scheduler.step()
    avg_train = epoch_loss / len(train_loader)
    train_losses.append(avg_train)

    if (epoch + 1) % 5 == 0:
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for imgs, _ in tqdm(val_loader, desc="  Validating", ncols=100, leave=False):
                imgs = imgs.to(device, non_blocking=True)
                with autocast():
                    loss, _, _ = model(imgs)
                    loss       = loss.mean()
                val_loss += loss.item()

        avg_val = val_loss / len(val_loader)
        val_losses.append((epoch + 1, avg_val))

        if avg_val < best_val:
            best_val = avg_val
            ckpt     = model.module if num_gpus > 1 else model
            torch.save(ckpt.state_dict(), os.path.join(CKPT_DIR, 'mae_best.pth'))

        print(f"Epoch [{epoch+1:>3}/{EPOCHS}]  Train Loss: {avg_train:.4f}  Val Loss: {avg_val:.4f}  LR: {scheduler.get_last_lr()[0]:.6f}  {'* saved' if avg_val == best_val else ''}")

    else:
        print(f"Epoch [{epoch+1:>3}/{EPOCHS}]  Train Loss: {avg_train:.4f}  LR: {scheduler.get_last_lr()[0]:.6f}")

torch.save((model.module if num_gpus > 1 else model).state_dict(), os.path.join(CKPT_DIR, 'mae_final.pth'))
print(f"\nTraining complete. Best Val Loss: {best_val:.4f}")
```

```python
plt.figure(figsize=(10, 4))
plt.plot(range(1, EPOCHS + 1), train_losses, label='Train Loss', linewidth=1.5)

val_epochs = [v[0] for v in val_losses]
val_vals   = [v[1] for v in val_losses]
plt.plot(val_epochs, val_vals, label='Val Loss', linewidth=1.5, marker='o', markersize=3)

plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('MAE Reconstruction Loss')
plt.legend()
plt.tight_layout()
plt.savefig('/kaggle/working/loss_curve.png', dpi=150)
plt.show()
```

## 3) Visualization Module

```python
best_model = MAE().to(device)
best_model.load_state_dict(torch.load(os.path.join(CKPT_DIR, 'mae_best.pth'), map_location=device))
best_model.eval()

def denormalize(tensor):
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1).to(tensor.device)
    std  = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1).to(tensor.device)
    return torch.clamp(tensor * std + mean, 0, 1)

def make_masked_image(img, mask):
    patches      = patchify(img.unsqueeze(0))
    mask_exp     = mask.unsqueeze(0).unsqueeze(-1).expand_as(patches)
    patches_vis  = patches.clone()
    patches_vis[mask_exp == 1] = 0.0
    return unpatchify(patches_vis).squeeze(0)

def unnormalize_patches(dec_out, imgs):
    # dec_out is [B, N, patch_dim]
    target = patchify(imgs)
    mean = target.mean(dim=-1, keepdim=True)
    var = target.var(dim=-1, unbiased=False, keepdim=True)
    return dec_out * (var + 1.e-6)**0.5 + mean
```

```python
samples, _ = next(iter(val_loader))
samples    = samples[:5].to(device)

with torch.no_grad():
    _, dec_out, mask = best_model(samples)

dec_out = unnormalize_patches(dec_out, samples)

fig, axes = plt.subplots(5, 3, figsize=(12, 18))
cols      = ['Masked Input', 'Reconstruction', 'Ground Truth']

for i in range(5):
    original      = denormalize(samples[i])
    masked_img    = denormalize(make_masked_image(samples[i], mask[i]))
    recon_patches = dec_out[i].unsqueeze(0)
    recon_img     = denormalize(unpatchify(recon_patches).squeeze(0))

    for ax, img, col in zip(axes[i], [masked_img, recon_img, original], cols):
        ax.imshow(img.cpu().permute(1, 2, 0).numpy())
        ax.axis('off')
        if i == 0:
            ax.set_title(col, fontsize=13, fontweight='bold', pad=8)

plt.suptitle('MAE Reconstruction Results', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('/kaggle/working/reconstructions.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: reconstructions.png")
```

## 9) PSNR & SSIM Evaluation

```python
from skimage.metrics import peak_signal_noise_ratio as psnr_fn
from skimage.metrics import structural_similarity as ssim_fn
best_model.eval()

all_psnr = []
all_ssim = []

with torch.no_grad():
    for imgs, _ in tqdm(val_loader, desc="Evaluating", ncols=100):
        imgs = imgs.to(device)

        with autocast():
            _, dec_out, mask = best_model(imgs)

        dec_out = unnormalize_patches(dec_out, imgs)
            
        for i in range(imgs.size(0)):
            original = denormalize(imgs[i]).cpu().permute(1, 2, 0).numpy()
            recon    = denormalize(unpatchify(dec_out[i].unsqueeze(0)).squeeze(0)).cpu().permute(1, 2, 0).numpy()

            original = np.clip(original, 0, 1)
            recon    = np.clip(recon,    0, 1)

            all_psnr.append(psnr_fn(original, recon, data_range=1.0))
            all_ssim.append(ssim_fn(original, recon, data_range=1.0, channel_axis=2))

mean_psnr = np.mean(all_psnr)
mean_ssim = np.mean(all_ssim)
std_psnr  = np.std(all_psnr)
std_ssim  = np.std(all_ssim)

print(f"PSNR : {mean_psnr:.2f} dB  ± {std_psnr:.2f}")
print(f"SSIM : {mean_ssim:.4f}     ± {std_ssim:.4f}")
```
